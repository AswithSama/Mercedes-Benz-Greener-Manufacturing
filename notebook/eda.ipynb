```python
# Get columns with categorical data types
categorical_columns = df2.select_dtypes(include=['object', 'category']).columns

```

```python
def catFeatureDist(dataframe,feature):
    plt.figure(figsize=(10,6))
    palette=sns.color_palette("husl",len(dataframe[feature].unique().tolist()))
    sns.stripplot(x=feature,y='y',data=dataframe,palette=palette)
    plt.title(f"{feature} category distribution")
    plt.show()
```

```python
for i in range(len(categorical_columns)):
    catFeatureDist(df2,categorical_columns[i])
```

```python
df3=df2.drop(['X4'],axis='columns')
```

```python
df3.head()
```

```python
#DECIDED TO YOU INTERQUARTILE RANGE TECHNIQUE TO REMOVE OUTLIERS


#upper and lower quartiles
q1=df3['y'].quantile(0.25)
q3=df3['y'].quantile(0.75)

#upper and lower limit
upper_limit=q3+(q3-q1)*1.5
lower_limit=q1-(q3-q1)*1.5

#filtering the outliers
outlier_indices=df3[(df3['y']>upper_limit) | (df3['y'] < lower_limit)].index

```

```python
#RESETTING THE INDEX IS REQUIRED AFTER DROPPING THE ROWS TO ENSURE SEQUENTIAL ORDER.
df4=df3.drop(outlier_indices,axis=0).reset_index(drop=True)  
df4.shape
```

```python
df4.head()
```

```python
X=df4.drop(['y'],axis=1)
X.head()
y=df4.y
```

```python
#REMOVING THE COLUMNS WHICH ARE ONE CATEOGORY DOMINANT AS THEY DONOT HAVE ANY IMPACT ON THE FINAL RESULT AND CERTAINLY AN OVERHEAD
check_X=X.select_dtypes(exclude='object')
remove_column=[]
for col in check_X:
    zeroes=(check_X[col]==0).sum()
    ones=(check_X[col]==1).sum()
    zeroes_percentage=(zeroes/(ones+zeroes))*100
    if zeroes_percentage >99 or zeroes_percentage<1:
        remove_column.append(col)
print(len(remove_column))
```

```python
X=X.drop(remove_column,axis=1)
```

```python
X.head()
```

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

# Combine the datasets
whole_dataset = pd.concat([X_train, X_test], axis=0)

# Define the categorical columns
categorical_columns = whole_dataset.select_dtypes(include='object').columns.tolist()

# Define the categorical pipeline with OneHotEncoder
cat_pipeline = Pipeline(steps=[
    ("one_hot_encoder", OneHotEncoder(handle_unknown='ignore'))
])

# Create a ColumnTransformer to handle the categorical pipeline
preprocessor = ColumnTransformer(transformers=[
    ("cat_pipeline", cat_pipeline, categorical_columns)
])

# Fit on the whole dataset to learn all categories
preprocessor.fit(whole_dataset)

# Transform the train and test sets
X_train_preprocessed = preprocessor.transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)

print(X_train_preprocessed.shape)
print(X_test_preprocessed.shape)
```
